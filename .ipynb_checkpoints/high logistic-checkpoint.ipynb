```python
new_tot_df
```

```python
whole_label = new_tot_df.label

whole_label

new_whole_label=whole_label.replace({ 'normal' : 0, 'neptune' : 1 ,'back': 1, 'land': 1, 'pod': 1, 'smurf': 1, 'teardrop': 1,'mailbomb': 1, 'apache2': 1, 'processtable': 1, 'udpstorm': 1, 'worm': 1,
                           'ipsweep' : 2,'nmap' : 2,'portsweep' : 2,'satan' : 2,'mscan' : 2,'saint' : 2
                           ,'ftp_write': 3,'guess_passwd': 3,'imap': 3,'multihop': 3,'phf': 3,'spy': 3,'warezclient': 3,'warezmaster': 3,'sendmail': 3,'named': 3,'snmpgetattack': 3,'snmpguess': 3,'xlock': 3,'xsnoop': 3,'httptunnel': 3,
                           'buffer_overflow': 4,'loadmodule': 4,'perl': 4,'rootkit': 4,'ps': 4,'sqlattack': 4,'xterm': 4})

new_whole_label.shape
```

```python
whole_proc = new_tot_df.protocol_type 
whole_proc
```

```python
new_whole_proc = whole_proc.replace({ 'tcp' : 1, 'udp' : 2 ,'icmp': 3})
```

```python
new_whole_proc
```

```python
whole_flag = new_tot_df.flag



new_whole_flag=whole_flag.replace({ 'OTH': 1, 'REJ': 2, 'RSTO': 3, 'RSTOS0': 4, 'RSTR': 5, 'S0': 6, 'S1': 7, 'S2': 8, 'S3': 9,'SF': 10, 'SH': 11})

new_whole_flag.shape
```

```python
log_df = new_tot_df.copy()
```

```python
from sklearn import preprocessing 
```

```python
log_df
```

```python
ole_service = preprocessing.LabelEncoder()
```

```python
log_df['o_service'] = ole_service.fit_transform(log_df['service'])
```

```python
print(ole_service.classes_)
```

```python
log_df
```

```python
log_df = log_df.drop(['service'],axis='columns')
```

```python
log_df
```

```python
log_df['o_label']= new_whole_label
log_df['oag'] = new_whole_flag
log_df['oroc'] = new_whole_proc
```

```python
new_whole_flag
```

```python
log_df['oflag'] = new_whole_flag
```

```python
log_df
```

```python
log_df = log_df.drop(['protocol_type','flag'],axis='columns')
```

```python
log_df = log_df.drop(['oag'],axis='columns')
```

```python
inputs = log_df.drop(['o_label'],axis = 'columns')
target = log_df.o_label
```

```python
inputs
```

```python
target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs,target,train_size=0.7)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(max_iter=500)
```

```python
logreg.fit(X_train,y_train)
```

```python
logreg.score(X_test,y_test)
```
